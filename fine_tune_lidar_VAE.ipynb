```python
# Load data

lidar_data = load_lidar_data(num_images=56, input_h=1280, input_w=1920, downsample=4, threshold=5, transform_method="normalization_without_mask")
num_samples, height, width = lidar_data.shape
input_size_flatten = height * width
train_data_loader, _ = create_dataloader(input_lidar=lidar_data, batch_size=8)
test_data = load_lidar_data(num_images=6, input_h=1280, input_w=1920, downsample=4, threshold=1)
```

```python
model, losses = vae_train(train_data_loader, input_dim=input_size_flatten, lr=0.00001, epochs=200, latent_dim=256, hidden_dim=(1024, 512), loss_beta=0.05, loss_mask=True)
visualize_lidar_data(model, test_data, downsample=4, recon_height=height, recon_width=width, num_images2show=6)
```
